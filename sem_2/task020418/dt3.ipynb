```python
import pydot
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree, datasets
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
clf_gini = DecisionTreeClassifier(criterion='gini', random_state=100, max_depth=3, min_samples_leaf=5)
clf_gini.fit(X_train, y_train)
tree.export_graphviz(clf_gini, out_file='sem_2/data/tree_gini.dot')
(graph,) = pydot.graph_from_dot_file('sem_2/data/tree_gini.dot')
graph.write_png('sem_2/data/tree_gini.png')
```

```python
img = mpimg.imread('sem_2/data/tree_gini.png')
imgplot = plt.imshow(img)
plt.show()
```

```python
clf_entropy = DecisionTreeClassifier(criterion='entropy', random_state=100, max_depth=3, min_samples_leaf=5)
clf_entropy.fit(X_train, y_train)
tree.export_graphviz(clf_entropy, out_file='sem_2/data/tree_entropy.dot')
(graph,) = pydot.graph_from_dot_file('sem_2/data/tree_entropy.dot')
graph.write_png('sem_2/data/tree_entropy.png')
```

```python
img = mpimg.imread('sem_2/data/tree_entropy.png')
imgplot = plt.imshow(img)
plt.show()
```

```python
y_pred_gini = clf_gini.predict(X_test)
y_pred_information_gain = clf_entropy.predict(X_test)

print("Accuracy with entropy param is ", accuracy_score(y_test, y_pred_information_gain))
print("Accuracy with gini param is ", accuracy_score(y_test, y_pred_gini))
```
